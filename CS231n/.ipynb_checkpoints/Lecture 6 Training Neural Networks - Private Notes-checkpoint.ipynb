# 3. Weight Initialization
<img src='./Lesson pic/6-38.png'>
Q: What happen when W=0 init is used?<br>
- First Idea : Small random numbers<br>
(gaussian with zero mean and 1e-2 standard deviation)<br>

```python
    W=0.01 * np.random.randn(D,H)
```

<font color='red'>Works okay for small networks, but problems with deeper networks</font>

<font color='blue'>E.g. 10-layer net with 500 neurons on each layer, using tanh non-linearities, and initializing as described in last slide.</font>

````python
import numpy as np
import matplotlib.pyplot as plt

#assume some unit gaussian 10-D input data
D=np.random.randn(1000,500)
hidden_layer_sizes=[500]*10 # [500, 500, 500, 500, 500, 500, 500, 500, 500, 500]
nonlinearities=['tanh']*len(hidden_layer_sizes) # len(hidden_layer_sizes)=10

act={'relu':lambda x:np.maximum(0,x), 'tanh':lambda x:np.tanh(x)}
Hs={}
for i in range(len(hidden_layer_sizes)):
    X= D if i==0 else Hs[i-1] # input at this layer - init:x=D else x=Hs[i-1]
    fan_in = X.shape[1] #500
    fan_out = hidden_layer_sizes[i]
    W=np.random.randn(fan_in, fan_out)*0.01 #layer_initialization
    print('X.shape:',X.shape,'fan_in:',fan_in,'fan_out',fan_out,'W.shape:',W.shape)
    
    H=np.dot(X,W)
    H=act[nonlinearities[i]](H)
    Hs[i]=H #cache result on this layer
    print('H.shape:',H.shape,'Hs.keys():',Hs.keys())
````

````python
print('\ninput layer had mean %f and std %f\n' %(np.mean(D), np.std(D)))
layer_means = [np.mean(H) for i,H in Hs.items()]
layer_stds = [np.std(H) for i,H in Hs.items()]
for i,H in Hs.items():
    print('Hidden layer %d had mean %f and std %f' %(i+1, layer_means[i], layer_stds[i]))    
````

````python
import matplotlib as mpl
mpl.rcParams['axes.unicode_minus'] = False

plt.figure()
plt.subplot(121)
plt.plot(Hs.keys(), layer_means, 'ob-')
plt.title('layer_mean')
plt.subplot(122)
plt.plot(Hs.keys(), layer_stds, 'or-')
plt.title('layer_std')

plt.figure()
for i,H in Hs.items():
    plt.subplot(1,len(Hs), i+1)
    plt.hist(H.ravel(), 30, range=(-1,1)) #parameter : array, bins, range
plt.show()
````

````python
# W=np.random.randn(fan_in, fan_out)*1. (0.01에서 1로 변경)

import numpy as np
import matplotlib.pyplot as plt

#assume some unit gaussian 10-D input data
D=np.random.randn(1000,500)
hidden_layer_sizes=[500]*10 # [500, 500, 500, 500, 500, 500, 500, 500, 500, 500]
nonlinearities=['tanh']*len(hidden_layer_sizes) # len(hidden_layer_sizes)=10

act={'relu':lambda x:np.maximum(0,x), 'tanh':lambda x:np.tanh(x)}
Hs={}
for i in range(len(hidden_layer_sizes)):
    X= D if i==0 else Hs[i-1] # input at this layer - init:x=D else x=Hs[i-1]
    fan_in = X.shape[1] #500
    fan_out = hidden_layer_sizes[i]
    W=np.random.randn(fan_in, fan_out)*1. #layer_initialization
    
    H=np.dot(X,W)
    H=act[nonlinearities[i]](H)
    Hs[i]=H #cache result on this layer
    
print('\ninput layer had mean %f and std %f\n' %(np.mean(D), np.std(D)))
layer_means = [np.mean(H) for i,H in Hs.items()]
layer_stds = [np.std(H) for i,H in Hs.items()]
for i,H in Hs.items():
    print('Hidden layer %d had mean %f and std %f' %(i+1, layer_means[i], layer_stds[i]))   
    
import matplotlib as mpl
mpl.rcParams['axes.unicode_minus'] = False

plt.figure()
plt.subplot(121)
plt.plot(Hs.keys(), layer_means, 'ob-')
plt.title('layer_mean')
plt.subplot(122)
plt.plot(Hs.keys(), layer_stds, 'or-')
plt.title('layer_std')

plt.figure()
for i,H in Hs.items():
    plt.subplot(1,len(Hs), i+1)
    plt.hist(H.ravel(), 30, range=(-1,1)) #parameter : array, bins, range
plt.show()
````

<font size=4 color='red'>Almost all neurons completely saturated, either -1 and 1. Gradients will be all zero.</font>

````python
# W=np.random.randn(fan_in, fan_out)/ np.sqrt(fan_in) "Xavier initialization"

import numpy as np
import matplotlib.pyplot as plt

#assume some unit gaussian 10-D input data
D=np.random.randn(1000,500)
hidden_layer_sizes=[500]*10 # [500, 500, 500, 500, 500, 500, 500, 500, 500, 500]
nonlinearities=['tanh']*len(hidden_layer_sizes) # len(hidden_layer_sizes)=10

act={'relu':lambda x:np.maximum(0,x), 'tanh':lambda x:np.tanh(x)}
Hs={}
for i in range(len(hidden_layer_sizes)):
    X= D if i==0 else Hs[i-1] # input at this layer - init:x=D else x=Hs[i-1]
    fan_in = X.shape[1] #500
    fan_out = hidden_layer_sizes[i]
    W=np.random.randn(fan_in, fan_out)/ np.sqrt(fan_in) #layer_initialization
    
    H=np.dot(X,W)
    H=act[nonlinearities[i]](H)
    Hs[i]=H #cache result on this layer
    
print('\ninput layer had mean %f and std %f\n' %(np.mean(D), np.std(D)))
layer_means = [np.mean(H) for i,H in Hs.items()]
layer_stds = [np.std(H) for i,H in Hs.items()]
for i,H in Hs.items():
    print('Hidden layer %d had mean %f and std %f' %(i+1, layer_means[i], layer_stds[i]))   
    
import matplotlib as mpl
mpl.rcParams['axes.unicode_minus'] = False

plt.figure()
plt.subplot(121)
plt.plot(Hs.keys(), layer_means, 'ob-')
plt.title('layer_mean')
plt.subplot(122)
plt.plot(Hs.keys(), layer_stds, 'or-')
plt.title('layer_std')

plt.figure()
for i,H in Hs.items():
    plt.subplot(1,len(Hs), i+1)
    plt.hist(H.ravel(), 30, range=(-1,1)) #parameter : array, bins, range
plt.show()
````
