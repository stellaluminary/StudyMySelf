# Titanic train Data
- PassengerID : 순번
- survival    : 생존여부(0:사망, 1:생존)
- pclass      : Ticket Class(사회지위) 1=1st, 2=2nd, 3=3rd
- sex         : 성별 (male, female)
- age         : 나이
- sibsp       : 탑승한 형제/배우자 수
- parch       : 탑승한 부모/자녀 수
- ticket      : 티켓 숫자
- fare        : 운임 요금
- cabin       : 승무원 번호
- embarked    : 승선항(C=Cherbourg, Q=Queenstown, S=Southamton)


# 가설
# Y = X1xW1+X2xW2...XnxWn

```python
import numpy as np
import pandas
import tensorflow as tf
```

```python
train_csv_data_p = pandas.read_csv('./Titanic/train.csv')
test_csv_data_p = pandas.read_csv('./Titanic/test.csv')
test_csv_sub_p = pandas.read_csv('./Titanic/gender_submission.csv')
```

```python
train_csv_data_p
```

```python
train_csv_data = train_csv_data_p.as_matrix()
test_csv_data = test_csv_data_p.as_matrix()
test_csv_sub = test_csv_sub_p.as_matrix()
train_csv_data
```

```python
# 데이터 전처리
# male -> 1
# female -> 0
for i in range(len(train_csv_data)):
    if train_csv_data[i, 4] == 'male':
        train_csv_data[i, 4] = 1
    else:                      # female
        train_csv_data[i, 4] = 0
        
for i in range(len(test_csv_data)):
    if test_csv_data[i, 3] == 'male':
        test_csv_data[i, 3] = 1
    else:
        test_csv_data[i, 3] = 0
```

```python
# 승선항 전처리
# S -> 1
# C -> 2
# Q -> 3
# nan -> 0    판단 못함, 공백
for i in range(len(train_csv_data)):
    if train_csv_data[i, 11] == 'S':
        train_csv_data[i, 11] = 1
    elif train_csv_data[i, 11] == 'C':
        train_csv_data[i, 11] = 2
    elif train_csv_data[i, 11] == 'Q':
        train_csv_data[i, 11] = 3
    if np.isnan(train_csv_data[i, 11]):
        train_csv_data[i, 11] = 0
                
for i in range(len(test_csv_data)):
    if test_csv_data[i, 10] == 'S':
        test_csv_data[i, 10] = 1
    elif test_csv_data[i, 10] == 'C':
        test_csv_data[i, 10] = 2
    elif test_csv_data[i, 10] == 'Q':
        test_csv_data[i, 10] = 3
    if np.isnan(test_csv_data[i, 10]):
        test_csv_data[i, 10] = 0
```

```python
X_PassengerData = train_csv_data[:, [2, # Pclass   
                                     4, # Sex
                                     6, # SibSp
                                     7, # Parch
                                     11 # Embarked
                                    ]]
Y_Survived = train_csv_data[:, 1:2]    # Survived
Test_X_PassengerData = test_csv_data[:, [1, # Pclass   
                                         3, # Sex
                                         5, # SibSp
                                         6, # Parch
                                         10 # Embarked
                                        ]]
Test_Y_Survived = test_csv_sub[:, 1:2]    # Survived
```

```python
# placeholder
X = tf.placeholder(tf.float32, shape=[None, 5])
Y = tf.placeholder(tf.float32, shape=[None, 1])
```

```python
# Variable
W = tf.Variable(tf.random_normal([5, 1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')
```

```python
# hypothesis
hypothesis = tf.sigmoid(tf.matmul(X, W) + b)
```

```python
# cost, loss 결과값과 예측값의 차이 : 작게 만드는 것이 학습
# 제곱오차
# Cross_Entropy
# Y=1, hypothesis = 1     정답   0에 가까워지고
# Y=1, hypothesis = 0     오답   무한대에 가까워진다
# Y=0, hypothesis = 1     오답   무한대에 가까워진다
# Y=0, hypothesis = 0     정답   0에 가까워지고
cost = -tf.reduce_mean(Y * tf.log(hypothesis) + 
                       (1 - Y) * tf.log(1-hypothesis))
```

```python
# Optimizer
train = tf.train.GradientDescentOptimizer(0.1).minimize(cost)
```

```python
predicted = tf.cast(hypothesis > 0.5, dtype=tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y),
                                 dtype=tf.float32))
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for step in range(10000):
        cost_val, _ = sess.run([cost, train], 
                              feed_dict={X:X_PassengerData,
                                        Y:Y_Survived})
          
        if step % 500 == 0:
            print("Step = ", step, ", Cost = ", cost_val)
            
    # 훈련데이터로 확인
    h, c, a = sess.run([hypothesis, predicted, accuracy],
                      feed_dict={X:X_PassengerData, 
                                Y:Y_Survived})
    print("\nAccuracy: ", a)

    # 테스트데이터로 확인
    print("\n Test CSV runningResult")
    h2, c2, a2 = sess.run([hypothesis, predicted, accuracy],
                         feed_dict={X:Test_X_PassengerData,
                                   Y:Test_Y_Survived})
    print("\nAccuracy: ", a2)
        
```
