```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
x=np.linspace(0,10,100)
y_cos=np.cos(x)
y_sin=np.sin(x)
y_sin2=np.sin(2*x)
mpl.rcParams['axes.prop_cycle']=mpl.cycler('color',['green','yellow','black'])
```

```python
plt.figure()
plt.plot(x,y_cos,'*')
plt.plot(x,y_sin)
plt.plot(x,y_sin2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('title')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y_cos,'g.')
plt.title('cos')
plt.subplot(1,2,2)
plt.plot(x,y_sin,'b-')
plt.title('sin')
plt.show()
```
