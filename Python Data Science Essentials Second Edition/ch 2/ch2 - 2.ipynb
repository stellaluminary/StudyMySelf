```python
def batch_read(filename, batch=5):
    with open(filename, 'rt') as data_stream:
        batch_output=list()
        for n, row in enumerate(csv.reader(data_stream,dialect='excel')):
            if n>0 and n%batch ==0:
                yield(np.array(batch_output))
                batch_output=list()
            batch_output.append(row)
        yield(np.array(batch_output))
```
